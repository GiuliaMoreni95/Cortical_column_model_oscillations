```python
from brian2 import *
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
#These are the values after the STDP conditioning
```

```python
#I import the files (these files were exported from the simulation code of the STDP conditioned network)
folder='../simulations_data/weights/'
values=[folder+'W_0.npy',folder+'W_8.npy',folder+'W_16.npy',folder+'W_80.npy',folder+'W_88.npy',folder+'W_96.npy',folder+'W_104.npy',folder+'W_160.npy',folder+'W_168.npy',folder+'W_176.npy',folder+'W_184.npy',folder+'W_256.npy',folder+'W_264.npy']

#Initialize arrays
means_new=[]
histo_all_new=[]
maxes_new=[]
lenghts=[]
all_conn=[]

for i in values:
    esp=np.array(np.load(i)) #array containing the weights of one pairs W_0 or W_8
    lenghts.append(len(esp)) #number of connections in each (W_0)
    histo_all_new += list(esp) #I add the array to the one containing all the connections strenghts

    all_conn.append(esp) #contains arrays, one for each pair 
    
    maxes_new.append(np.max(esp)) #array with max values for each pair group of connections 
    mean_val=np.mean(esp,dtype=np.float64)  #mean value for each pair group of connections
    means_new.append(mean_val) #array with mean values for each pair group of connections 
    
    #I plot the histogram of weights
    plt.hist(esp,bins=100)
    plt.ylim(0,5)
```

```python
#I AM DOING SOME CHECKS 
print(len(histo_all_new)) #contains the total number of connections 
print(lenghts)#contains the number of connections in each pair of populations 
print(sum(lenghts)) #This should be same as len(histo_all_new).
#GREAT IT IS THE SAME, I am doing things right
```

```python
#MAX value of the connections
print(np.max(histo_all_new))
#MAX value in each group of connections 
print(maxes_new)
```

```python
#I import the files (these files were exported from the simulation code)
folder='../simulations_data/weights/' 
values=[folder+'W_0_in.npy',folder+'W_8_in.npy',folder+'W_16_in.npy',folder+'W_80_in.npy',folder+'W_88_in.npy',folder+'W_96_in.npy',folder+'W_104_in.npy',folder+'W_160_in.npy',folder+'W_168_in.npy',folder+'W_176_in.npy',folder+'W_184_in.npy',folder+'W_256_in.npy',folder+'W_264_in.npy']
#Initialize arrays
means_old=[]
histo_all_old=[]
maxes_old=[]
all_conn_old=[]
lenghts_old=[]

for i in values:
    esp=np.array(np.load(i)) #array containing the weights of one pairs W_0_in or W_8_in
    histo_all_old += list(esp) #I add the array to the one containing all the connections strenghts
    lenghts_old.append(len(esp)) #number of connections in each (W_0)

    all_conn_old.append(esp) #contains arrays, one for each pair 
    
    maxes_old.append(np.max(esp)) #array with max values for each pair group of connections 
    mean_val=np.mean(esp,dtype=np.float64)  #mean value for each pair group of connections
    means_old.append(mean_val) #array with mean values for each pair group of connections 
        
    #I plot the histogram
    plt.hist(esp,bins=100)
    #plt.ylim(0,5)
```

```python
#I AM DOING SOME CHECKS 
print(len(histo_all_old)) #contains the total number of connections 
print(lenghts_old)#contains the number of connections in each pair of populations 
print(sum(lenghts_old)) #This should be same as len(histo_all_new).
#GREAT IT IS THE SAME, I am doing things right
```

```python
#MAX value of the connections
print(np.max(histo_all_old))
#MAX value in each group of connections 
print(maxes_old)
```

```python
#MEAN OF ALL THE VALUES ALL TOGHTER and then compute percentage change from those means
#I want to know how much average increase between NAIVE network and STDP contitioned
mn_all_old=np.mean(histo_all_old)
mn_all_new=np.mean(histo_all_new)
print(mn_all_old)
print(mn_all_new)

perc_change_tot=(mn_all_new-mn_all_old)/mn_all_old
print('mean percentage change')
print(perc_change_tot)
```

```python
values=[folder+'W_0_in.npy',folder+'W_8_in.npy',folder+'W_16_in.npy',folder+'W_80_in.npy',folder+'W_88_in.npy',folder+'W_96_in.npy',folder+'W_104_in.npy',folder+'W_160_in.npy',folder+'W_168_in.npy',folder+'W_176_in.npy',folder+'W_184_in.npy',folder+'W_256_in.npy',folder+'W_264_in.npy']

#Initialize arrays
means_old_INCR=[]
histo_all_INCR=[]
maxes_INCR=[]
all_conn_INCR=[]

for i in values:

    esp=np.array(np.load(i))+perc_change_tot*np.array(np.load(i))
    histo_all_INCR += list(esp)
    all_conn_INCR.append(esp)    
    
    mean_val=np.mean(esp)
    means_old_INCR.append(mean_val)
    
    maxes_INCR.append(np.max(esp))
    
    plt.hist(esp,bins=100)
```

```python
#MEAN OF EACH PAIR and then compute percentage change from each pair
perc_change_INC=(np.array(means_old_INCR) - np.array(means_old))/np.array(means_old)
print(perc_change_INC)
print(np.mean(perc_change_INC))
#OK CHECK WORKED *check that I was really incresing by 0.70 as desired
```

```python
#MAX value of the connections
print(np.max(histo_all_INCR))
#MAX value in each group of connections 
print(maxes_INCR)
```

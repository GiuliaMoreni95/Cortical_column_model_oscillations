```python
#Create fig
fig,ax = plt.subplots(figsize =(10, 10))

#data to plot
sy = [
s_vip6,
s_sst6,
s_pv6,
s_e6,
s_vip5,
s_sst5,
s_pv5,
s_e5,
s_vip4,
s_sst4,
s_pv4,
s_e4,
s_vip23,
s_sst23,
s_pv23,
s_e23,
]

labels=['v6', 's6',
                    'p6', 'e6','v5', 's5',
                    'p5', 'e5','v4', 's4',
                    'p4', 'e4','v23', 's23',
                    'p23', 'e23']


colors = ['orange','green', 'royalblue', 'crimson','orange','green', 'royalblue', 'crimson','orange','green', 'royalblue', 'crimson','orange','green', 'royalblue', 'crimson','orange']


y_pos = np.arange(len(labels))
width=0.5
alpha_me=0.7
# Create bars
plt.barh(y_pos[0:4],sy[0:4],width,color=colors[0:4],alpha=alpha_me)
plt.barh(y_pos[4:8],sy[4:8],width,color=colors[4:8],alpha=alpha_me)
plt.barh(y_pos[8:12],sy[8:12],width,color=colors[8:12],alpha=alpha_me)
plt.barh(y_pos[12:16],sy[12:16],width,color=colors[12:16],alpha=alpha_me)

#no borders of the plot 
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
#x and y axis thick 
plt.setp(ax.spines.values(), linewidth=3)

rect6=mpatches.Rectangle((0,-0.8),0.35,4.3, 
                    #fill = False,
                        alpha=0.1,
                   facecolor="grey",
                    #color = "black",
                    linewidth = 1)

rect5=mpatches.Rectangle((0,3.5),0.35,4, 
                    fill = False,
                        alpha=0.1,
                   #facecolor="grey",
                    color = "black",
                    linewidth = 1)

rect4=mpatches.Rectangle((0,7.5),0.35,4, 
                    #fill = False,
                        alpha=0.1,
                   facecolor="grey",
                    #color = "red",
                    linewidth = 1)


rect23=mpatches.Rectangle((0,11.5),0.35,4, 
                fill = False,
                    alpha=0.1,
               #facecolor="grey",
                color = "gray",
                linewidth = 1)

#I add the rectangles
plt.gca().add_patch(rect6)
plt.gca().add_patch(rect5)
plt.gca().add_patch(rect4)
plt.gca().add_patch(rect23)

#To put layers on the y axis
lab = ('L6', 'L5',
                'L4', 'L2/3')
y_pos = [2,6,10,14]
plt.yticks(y_pos, lab,size=35)
#plt.ylabel('Populations',size=20)

# Taylor the x-axis
plt.xlabel('Synchrony',size=40)
plt.locator_params(axis='x', nbins=4)
ax.xaxis.set_tick_params(width=3)
plt.xticks(size=35)

ax.tick_params(top=False,
               bottom=True,
               left=False,
               right=False,
               labelleft=True,#False (to remove layers from y axis)
               labelbottom=True)


legend_elements = [
               Patch(facecolor='crimson',alpha=alpha_me,
                     label='E'),
                   Patch(facecolor='royalblue',alpha=alpha_me,
                     label='PV'),
               Patch(facecolor='green',alpha=alpha_me, 
                     label='SST'),
                Patch(facecolor='orange',alpha=alpha_me, 
                     label='VIP')]

#ax.legend(handles=legend_elements, loc='lower right',fontsize=15)

#Save fig
plt.savefig('IMAGES/Fig2D_synch.png')
# Show graphic
plt.show()
```

```python
#IS THIS worth noting? 
#I am averaging the result from the 3 inhibitory neurons
#to get an mean inhibition value for each layer
s_in23=(s_pv23+s_sst23+s_vip23)/3
s_in4= (s_pv4+s_sst4+s_vip4)/3
s_in5= (s_pv5+s_sst5+s_vip5)/3
s_in6= (s_pv6+s_sst6+s_vip6)/3

#I obtain the exact same trend as in the paper of Diesmann as shown by this graph. 
fig = plt.figure(figsize =(10, 7))
syn=[
    s_e23,
    s_e4,
    s_e5,
    s_e6,
    s_in23,
    s_in4,
    s_in5,
    s_in6,
]
bars = ('e23', 'e4',
                    'e5', 'e6','in23', 'in4',
                    'in5', 'in6')
y_pos = np.arange(len(bars))

# Create bars
plt.barh(y_pos,syn,color='c')

# Create names on the x-axis
plt.yticks(y_pos, bars)
plt.xlabel('Synchrony',size=15)
plt.ylabel('Populations',size=15)

plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
# Show graphic
plt.show()
```

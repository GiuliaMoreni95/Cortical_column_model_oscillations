```python
#Raster plot of the cortical column
```

```python
Ntot=5000 #total number of neurons (rows) in the plot
#Need this for the raster plot
N1=int(0.0192574218*Ntot) 
N2_3=int(0.291088453*Ntot)
N4=int(0.237625904*Ntot)
N5=int(0.17425693*Ntot)
N6= Ntot-N2_3-N4-N5

#FROM LAYER 2/3 TOP to LAYER 6 bottom of the plot
#I need the indexes to build the rows
i_e23=S_e23i+ N[0][3]+N[0][2]+N[0][1]
i_pv23=S_pv23i+ N[0][3]+N[0][2]
i_sst23=S_sst23i+ N[0][3]
i_vip23=S_vip23i

i_e4=S_e4i+ N[1][3]+N[1][2]+N[1][1]
i_pv4=S_pv4i+ N[1][3]+N[1][2]
i_sst4=S_sst4i+ N[1][3]
i_vip4=S_vip4i

i_e5=S_e5i+ N[2][3]+N[2][2]+N[2][1]
i_pv5=S_pv5i+ N[2][3]+N[2][2]
i_sst5=S_sst5i+ N[2][3]
i_vip5=S_vip5i

i_e6=S_e6i+ N[3][3]+N[3][2]+N[3][1]
i_pv6=S_pv6i+ N[3][3]+N[3][2]
i_sst6=S_sst6i+ N[3][3]
i_vip6=S_vip6i

#fig, ax = plt.subplots(figsize =(7, 17)) #Tight
#fig, ax = plt.subplots(figsize =(9, 16)) 
fig, ax = plt.subplots(figsize =(10, 16)) #Large

#Layer 2/3
plt.plot(S_e23t,i_e23+N6+N5+N4,'.', markersize=3,color='r', label='E')
plt.plot(S_pv23t,i_pv23+N6+N5+N4,'.', markersize=3,color='b', label='PV')
plt.plot(S_sst23t,i_sst23+N6+N5+N4,'.', markersize=3,color='g', label='SST')
plt.plot(S_vip23t,i_vip23+N6+N5+N4,'.', markersize=3,color='darkorange', label='VIP')
#Layer 4
plt.plot(S_e4t,i_e4+N6+N5,'.', markersize=3,color='r')
plt.plot(S_pv4t,i_pv4+N6+N5,'.', markersize=3,color='b')
plt.plot(S_sst4t,i_sst4+N6+N5,'.', markersize=3,color='g')
plt.plot(S_vip4t,i_vip4+N6+N5,'.', markersize=4,color='darkorange')
#Layer 5
plt.plot(S_e5t,i_e5+N6,'.', markersize=3,color='r')
plt.plot(S_pv5t,i_pv5+N6,'.', markersize=3,color='b')
plt.plot(S_sst5t,i_sst5+N6,'.', markersize=3,color='g')
plt.plot(S_vip5t,i_vip5+N6,'.', markersize=4,color='darkorange')
#Layer 6
plt.plot(S_e6t,i_e6,'.', markersize=3,color='r')
plt.plot(S_pv6t,i_pv6,'.', markersize=3,color='b')
plt.plot(S_sst6t,i_sst6,'.', markersize=3,color='g')
plt.plot(S_vip6t,i_vip6,'.', markersize=4,color='darkorange')

plt.ylim(-30,5205)
plt.xlim(0,1500)
#plt.legend(fontsize=19) #If you want to plot the standard legend


#Layout
plt.setp(ax.spines.values(), linewidth=3)
#Remove only 2 edges of the box
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')

#Ticks size and lables
plt.xticks(size=30)
plt.locator_params(axis='x', nbins=5) #How many ticks on x axis
ax.xaxis.set_tick_params(width=3)
ax.yaxis.set_tick_params(width=3)
plt.xlabel('Time (ms)', size=35)
#plt.ylabel('neuron index', size=30)

#Create names of layers on the y-axis
lab = ('L6', 'L5',
               'L4', 'L2/3')
y_pos = [900,2000,3040,4400] #position where to put the labels
plt.yticks(y_pos, lab,size=28)
plt.yticks(size=30)


#Legend with taylored sizes
l1= plt.legend(loc='upper center', bbox_to_anchor=(0.55, 1.07),
          ncol=2, fancybox=True, markerscale=7, shadow=False, fontsize=25)
l1.get_frame().set_linewidth(3.0)
l1.get_frame().set_edgecolor('k')

#Save the fig
plt.savefig('../IMAGES/Fig2A_raster_large.png')

#Show the plot
plt.show()
```

```python
#Data I will plot
data=[vip6_sp,sst6_sp,pv6_sp,e6_sp,vip5_sp,sst5_sp,pv5_sp,e5_sp,vip4_sp,sst4_sp,pv4_sp,e4_sp,vip23_sp,sst23_sp,pv23_sp,e23_sp]

#I create the figure
fig, ax = plt.subplots(figsize =(12, 9))

#Colors for each group
colors = ['orange','forestgreen', 'cornflowerblue', 'crimson','orange','forestgreen', 'cornflowerblue', 'crimson','orange','forestgreen', 'cornflowerblue', 'crimson','orange','forestgreen', 'cornflowerblue', '#E50000']

#features of the bar plot (tringle for mean, vertical black bar for median)
meanpointprops = dict(markeredgecolor='black',
                      markerfacecolor='black',markersize=15)

medianprops = dict(linestyle='solid', linewidth=2, color='black')

#Layout of the barplot
box1=ax.boxplot(data,vert=False,patch_artist=True,medianprops=medianprops
,meanprops=meanpointprops,showmeans=True)

#Layout of the figure
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.setp(ax.spines.values(), linewidth=2) #thickness of axes
ax.xaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5) #Grid 

#Labels on the axes and size of ticks
#plt.title('Firing rates (sp/s) of populations')
#plt.ylabel('Populations',fontsize=20)
plt.xlabel('Firing rates (spikes/s)',fontsize=35)
ax.xaxis.set_tick_params(width=3)
ax.yaxis.set_tick_params(width=3)
plt.xticks(size=30)

# Create names on the y-axis
lab = ('L6', 'L5',
                'L4', 'L2/3')
y_pos = [3,7,11,15]
plt.yticks(y_pos, lab,size=30)



#Assign colors to the box plot
for patch, color in zip(box1['boxes'], colors):
    patch.set_facecolor(color)


#CUSTOMIZE THE LEGEnD (If you need it uncomment below)
legend_elements = [
               Patch(facecolor='crimson', edgecolor='k',
                     label='E'),
                   Patch(facecolor='cornflowerblue', edgecolor='k',
                     label='PV'),
               Patch(facecolor='forestgreen', edgecolor='k',
                     label='SST'),
                Patch(facecolor='orange', edgecolor='k',
                     label='VIP')]

#ax.legend(handles=legend_elements, loc='lower right',fontsize=23) #Uncomment if you want the legend


plt.savefig('../IMAGES/Fig2C_var.png')
plt.show()
```
